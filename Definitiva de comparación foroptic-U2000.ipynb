```python
#Importa el módulo numpy
import numpy as np
#Importa el modulo de pandas y crea dataframes a partir de los usuarios del foroptic
import pandas as pd
#Clientes del servidor A
foroptic_A = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/clientes (A).xlsx')
#Clientes del servidor B
foroptic_B = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/clientes (B).xlsx')
```

```python
# Comparar ususarios de OLT con base de datos del servidor U2000 y agregar columna Operstatus(activo, deshabilitado)
def agregar_valor(row):
    matching_row = bdd_u2000[
        (bdd_u2000['Name'] == row['Contrato'])
    ]
    if not matching_row.empty:
        return matching_row.iloc[0][['OperStatus', 'RunStatus']]  # retornar la columna que desees
    return None

usuarios_GPON_Total[['OperStatus', 'RunStatus']] = usuarios_GPON_Total.apply(agregar_valor, axis=1)#Agregar columna
```

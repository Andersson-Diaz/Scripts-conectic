```python
#usuarios inscritos en la plataforma en la nube(Facturacion) en el servidor A
foroptic_A = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Ordenes de servicio 2023/clientes (A).xlsx')
#Clientes B
#usuarios inscritos en la plataforma en la nube(Facturacion) en el servidor B
foroptic_B = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Ordenes de servicio 2023/clientes (B).xlsx')
```

```python
#Une en un solo dataframe los dos foroptics (Sistemas de facturacion)
foroptic_A_y_B = pd.concat([foroptic_A, foroptic_B], ignore_index=True)
```

```python
foroptic_A_y_B['Contr']=foroptic_A_y_B['Contr'].astype(int)
```

```python
instalaciones = pd.concat([empaquetamiento_A,empaquetamiento_B,instalaciones_A,instalaciones_B], ignore_index=True)
instalaciones
```

```python
# Función para comprobar si las palabras clave están presentes en alguna de las dos columnas
def contiene_palabras_clave(row):
    palabras_clave_gpon = ['GPON', 'SN', 'DROP', 'ONU']
    palabras_clave_docsis = ['DOCSIS', 'SNR', 'MAC', 'MODEM']
    for palabra in palabras_clave_gpon:
        if palabra in row['Observac al crear'] or palabra in row['Observac al cerrar']:
            return 'GPON'
        else:
          for palabra in palabras_clave_docsis:
              if palabra in row['Observac al crear'] or palabra in row['Observac al cerrar']:
                  return 'DOCSIS'
    return ''
```

```python
# Aplicar la función a lo largo de las filas y crear una nueva columna 'resultado'
instalaciones['Tecnologia'] = instalaciones.apply(contiene_palabras_clave, axis=1)
```

```python
#Comparar por mac en las dos bases de datos y extraer la columna con la informacion de contrato
def agregar_valor(row):
    matching_row = foroptic_A_y_B[
        (foroptic_A_y_B['Contr'] == row['Contrato'])
    ]
    if not matching_row.empty:
        return matching_row.iloc[0]['Servicio']  # Cambiar por la columna que desees
    return None

instalaciones['Servicio en foroptic'] = instalaciones.apply(agregar_valor, axis=1)
```

```python
# Convertir la columna 'Fecha' al formato datetime
instalaciones['Fecha'] = pd.to_datetime(instalaciones['Fecha'])
instalaciones['Fecha Creac'] = pd.to_datetime(instalaciones['Fecha Creac'])
```

```python
# Crear una nueva columna 'DiferenciaTiempo' con la diferencia de tiempo entre las fechas
instalaciones['Tiempo de ejecucion (dias)'] = (instalaciones['Fecha'] - instalaciones['Fecha Creac']).dt.days
```

```python
# Crear una nueva columna 'NombreMes' con el nombre del mes
instalaciones['Mes de instalacion'] = instalaciones['Fecha'].dt.month_name()
```

```python
# Contar la frecuencia de cada valor único en la columna 'DiferenciaDias'
frecuencia_dias = instalaciones['Tiempo de ejecucion (dias)'].value_counts()

# Crear un gráfico de barras para cada día específico
fig, ax = plt.subplots()
bars = frecuencia_dias.plot(kind='bar', edgecolor='black', ax=ax)

# Agregar etiquetas de texto a cada barra con el valor exacto
for bar in bars.patches:
    yval = bar.get_height()
    xval = bar.get_x() + bar.get_width() / 2
    ax.text(xval, yval, round(yval, 2), ha='center', va='bottom', color='black', fontsize=8)

plt.title('Teimpo de ejecucion de las instalaciones')
plt.xlabel('Tiempo de ejecucion en días')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Contar la frecuencia de cada valor único en la columna 'DiferenciaDias'
frecuencia_meses = instalaciones['Mes de instalacion'].value_counts()
frecuencia_meses.sort_values()
# Crear un gráfico de barras para cada día específico
fig, ax = plt.subplots()
bars = frecuencia_meses.plot(kind='bar', edgecolor='black', ax=ax)
print(frecuencia_meses)
# Agregar etiquetas de texto a cada barra con el valor exacto
for bar in bars.patches:
    yval = bar.get_height()
    xval = bar.get_x() + bar.get_width() / 2
    ax.text(xval, yval, round(yval, 2), ha='center', va='bottom', color='black', fontsize=8)

plt.title('Numero de instalaciones por mes')
plt.xlabel('meses')
plt.ylabel('Frecuencia')
plt.show()
```
